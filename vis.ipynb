```python
# Imports, setup
import pandas as pd
import altair as alt
alt.data_transformers.disable_max_rows()

# Data, default theme
licenses = pd.read_csv("licenses_clean.csv")
licenses
licenses_melt = pd.melt(licenses, id_vars=["License"], value_vars=["Rate of Acceptance", "Rate of Acceptance for JIs"], var_name="Population", value_name="Acceptance Rate")
licenses_melt = licenses_melt.replace("Licensed", "", regex=True)
licenses_melt
```

```python
alt.Chart(licenses).mark_bar().encode(
    x="Acceptance Rate:Q",
    y="Population:N",
    color="Population:N",
    row=alt.Row("License")
).transform_fold(as_=['Population', 'Acceptance Rate'], fold=["Rate of Acceptance for JIs", "Rate of Acceptance"]).display()
```

```python
p1 = alt.Chart(licenses, title='Comparing License Acceptance Rates').mark_point().encode(
        alt.Y('License:O').title("License Type"),
        alt.X('Rate of Acceptance').title("Acceptance").axis(format='%'),
        color=alt.value("blue")
)

p2 = alt.Chart(licenses, title='Comparing License Acceptance Rates').mark_point().encode(
        alt.Y('License:O').title("License Type"),
        alt.X('Rate of Acceptance for JIs').title("Acceptance").axis(format='%'),
        color=alt.value("purple")
)

p1 + p2
```

```python
alt.Chart(licenses_melt).mark_bar().encode(
    x=alt.X("License:O",axis=alt.Axis(labelLimit=250)), 
    xOffset="Population:N",
    y=alt.Y("Acceptance Rate:Q").axis(format='%'),
    color=alt.Color("Population:N")
)
```
